# Soundscapy tutorials

[![Open In
Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/MitchellAcoustics/Soundscapy/blob/main/examples/_site/index.ipynb)

Each tutorial demonstrates a different aspect of soundscape analysis
with Soundscapy.

If you are new to soundscape analysis, start with [About Soundscape
Analysis](background.qmd) for an overview of the core concepts before
diving into code.

## Where to start

| I want to… | Go to |
|------------------------------------|------------------------------------|
| Understand the concepts behind the analysis | [About Soundscape Analysis](background.qmd) |
| Run something immediately | [Quick Start](QuickStart.qmd) |
| Learn step by step, new to Python | [Getting Started](0_QuickStart_for_Beginners.qmd) |
| See the analytical approach behind the library | [Analysing and Representing Soundscapes](HowToAnalyseAndRepresentSoundscapes.qmd) |
| Work through a full assessment step by step | Tutorials 1–6 below, in order |
| Analyse audio recordings | [Binaural Audio Analysis](BinauralAnalysis.qmd) — requires `soundscapy[audio]` |
| Use the Soundscape Perception Index or CircE | [Soundscape Perception Index](4_Understanding_Soundscape_Perception_Index.qmd) — requires `soundscapy[r]` |

## Tutorials

1.  [Understanding Soundscape
    Analysis](1_Understanding_Soundscape_Analysis.qmd) — ISO 12913
    framework, the circumplex model, and PAQ data
2.  [Working with Survey
    Data](2_Working_with_Soundscape_Survey_Data.qmd) — loading,
    validating, and transforming PAQ data
3.  [Visualization Techniques](3_Advanced_Visualization_Techniques.qmd)
    — scatter plots, density plots, joint plots, multi-panel figures
4.  [Soundscape Perception
    Index](4_Understanding_Soundscape_Perception_Index.qmd) — fitting
    MSN distributions and scoring against target distributions
5.  [Working with Databases](5_Working_with_Soundscape_Databases.qmd) —
    ISD, ARAUS, SATP, and language-specific coordinates
6.  [Complete Assessment Workflow](6_Soundscape_Assessment_Tutorial.qmd)
    — end-to-end assessment of a custom dataset
