```r
library(sn)
library(tidyverse)
```

```r
Sigma <- matrix(c(1, 0.4, 0.4, 0.6), 2, 2)
cpM <- list(mean=c(0,-1), var.cov=Sigma, gamma1=c(0, -0.6))
dpM <- cp2dp(cpM, family="SN")
yM <- rmsn(10, dp=dpM)

```

```r
yM
```

```r
Omega <- matrix(c(0.15, -0.058, -0.058, 0.093), 2, 2)
dpM <- list(xi=c(0.06, 0.597), Omega=Omega, alpha=c(0.868, -0.561))
yM <- rmsn(1000, dp=dpM)
yM
```
